```python
import pickle
import numpy as np
```

```python
## Code to load the array
with open('/root/Desktop/mlss18-assignment1/Assignment_1_1/array.pickle', 'rb') as p:
    arr = pickle.load(p)
   # print(arr)
```

Check that the size of the array (arr) is 1000. Please report if any discrepancy.

## Task 1: Resize the array to a (20x50) matrix

```python
## BEGIN CODE
np_arr=np.array(arr)
np_arr.resize(20,50)
#print(np_arr)
## END CODE
```

## Task 2: Find the transpose of this matrix

```python
## BEGIN CODE
np_arr1=np_arr.transpose()
#print(np_arr1)
## END CODE
```

## Task 3: Find the dot product of the matrices created in Tasks 1 & 2

```python
## BEGIN CODE
np_arr2=np.dot(np_arr,np_arr1)
print(np_arr2)
## END CODE
```

## Task 4: Find the row-wise sum of the resulting matrix

```python
## BEGIN CODE
import numpy as np

np_arr3=np.sum(np_arr2,axis=1)
print(np_arr3)
## END CODE
```

## Task 5: Find the square root of all elements in the resulting matrix

```python
## BEGIN CODE
np_arr4=np_arr**0.5
print(np_arr4)
## END CODE
```

## Task 6: Find the sine value of all elements in the resulting matrix

```python
## BEGIN CODE
np_arr5=np.sin(np_arr4)
print(np_arr5)
## END CODE
```
